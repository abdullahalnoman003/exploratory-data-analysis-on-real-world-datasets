```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./raw dataset/dataset.csv")
shape = df.shape
head = df.head(10)
tail = df.tail(10)
print(f"Dataset shape: {shape}")
print(f"First 10 rows of the dataset:\n{head}\n")
print(f"Last 10 rows of the dataset:\n{tail}\n")
df.info()
print(f"Number of duplicated rows: {df.duplicated().sum()}")

sample = df.sample(10, random_state=42)
print("Sample of 10 random rows from the dataset: \n", sample)
sum = df.isna().sum()
print(f"Number of missing values in each column: {sum}")
missing_percent = (df.isnull().sum()/len(df))*100
print(f"Percentage of missing values in each column: {missing_percent}")
print(f"Column names: {list(df.columns)}")
df.describe()
```
